**<h1> Introducción</h1>**

El Virus del Papiloma Humano (VPH) es una de las infecciones de transmisión sexual más comunes a nivel mundial, con más de 100 tipos diferentes, algunos de los cuales están estrechamente relacionados con el desarrollo del cáncer cervicouterino. Este tipo de cáncer se origina en las células del cuello uterino, la parte inferior del útero que conecta con la vagina, y representa una de las principales causas de mortalidad relacionada con el cáncer entre mujeres en varios países, especialmente en aquellos con acceso limitado a programas de cribado y vacunación contra el VPH.

La detección oportuna del VPH y las lesiones precancerosas en el cuello uterino es fundamental para prevenir el desarrollo del cáncer cervicouterino. La implementación de programas de vacunación contra el VPH y el cribado regular mediante pruebas de Papanicolaou (Pap) y pruebas de detección del VPH pueden identificar infecciones de alto riesgo y cambios precancerosos en las células cervicales antes de que se conviertan en cáncer. Estas estrategias de prevención y detección temprana han demostrado ser efectivas en la reducción significativa de la incidencia y mortalidad por cáncer cervicouterino.

A pesar de estos avances, el cáncer cervicouterino sigue siendo un problema de salud pública significativo, lo que subraya la importancia de la educación y concienciación sobre el VPH y el cáncer cervicouterino. El conocimiento y la comprensión de estos temas son esenciales para promover la participación en programas de vacunación y detección, así como para fomentar prácticas de salud sexual responsable.

```python
#Lidiando con valores atípicos de datos numéricos

def calcular_filtro_iqr(df, columna):
  """
  Calcula el IQR y los filtros inferior y superior para una columna.

  Parámetros:
    df: El DataFrame que contiene la columna.
    columna: El nombre de la columna numérica.

  Devuelve:
    filtro_inferior: Un objeto Series que indica los valores que no son outliers.
    filtro_superior: Un objeto Series que indica los valores que no son outliers.
  """
  iqr = df[columna].quantile(0.75) - df[columna].quantile(0.25)
  filtro_inferior = df[columna] > df[columna].quantile(0.25) - (iqr * 1.5)
  filtro_superior = df[columna] < df[columna].quantile(0.75) + (iqr * 1.5)
  return filtro_inferior, filtro_superior

```

En este caso, un valor atípico se define como un valor que se encuentra fuera del rango normal de la distribución de los datos numéricos.
Todos aquellos valores que se encuentran por debajo del umbral inferior o por encima del superior como mostró la mayoría de nuestros datos, se consideran valores atipicos.
Al observar la modificación de los datos, se ve que la media de cada una de las categorías se movio muy poco y que fueron pocos los valores atipicos eliminados.
